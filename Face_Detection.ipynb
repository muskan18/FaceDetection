```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
im=plt.imread("./download.jpg")
```

```python
plt.imshow(im)
```

```python
detector=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces=detector.detectMultiScale(im)
```

```python
for face in faces:
    x,y,w,h = face
    modified_img=cv2.rectangle(im,(x,y),(x+w,y+h),(255,255,255),2)    
```

```python
plt.imshow(modified_img)
```
